## Lektion 50. 2 Variablen  mit ggplot2 darstellen

In dieser Lektion werden wir uns kurze Beispiele dafür anschauen, wie wir zwei Variablen eines Datansatzes vergleichen können. Für diese Beispiele brauchen wir die Kontollmöglichkeiten, die ggplot bietet, können aber auch kurz anschauen wozu qplot in der Lage ist. Wir nutzen den Film-Datensatz aus ggplot2!

```r
library(ggplot2)
library(ggplot2movies)
df <- movies
```

## qplot()

```r
qplot(x=year, y=rating, data = df, geom = "density2d")
```

```r
pl <- ggplot(movies,aes(x = year,y=rating))
```

## 2d Bin Chart

```r
pl + geom_bin2d()
```

```r
# Größe einstellen
pl + geom_bin2d(binwidth=c(2,1))
```

## 2d Density Plot

```r
pl + geom_density2d()
```

## 2d Hex Plot

```r
# install.packages("hexbin")
```

```r
pl + geom_hex()
```

```r
pl + geom_hex() + scale_fill_gradient(high='red',low='blue')
```
