**Schreiben wir eine Custom-Function, die Stats-Summary ermöglicht. Weitere Details s. [hier](https://medium.com/@gscheithauer/how-to-add-number-of-observations-to-a-ggplot2-boxplot-b22710f7ef80)**

```r
# summary custom function
stat_box_data <- function(y, upper_limit = max(mtcars$mpg) * 1.15) {
  return( 
    data.frame(
      y = 0.95 * upper_limit,
      label = paste('Anzahl =', length(y), '\n',
                    'Mean (Mittelw.) =', round(mean(y), 1), '\n',
                    'Median (Mitte) =', round(median(y), 1), '\n')
    )
  )
}
```

```r
# stat-summary anzeigen 
pl <- ggplot(mtcars, aes(factor(cyl), mpg), axes=FALSE)
pl + geom_boxplot() +
  stat_summary(
    fun.data = stat_box_data, 
    geom = "text", 
    hjust = 0.5,
    vjust = 0.9
  )
```

```r
pl + geom_boxplot() + coord_flip()
```

```r
pl + geom_boxplot(aes(fill = factor(cyl)))
```

```r
pl + geom_boxplot(fill = "grey", color = "red")
```
