## Lektion 47. Scatterplots mit ggplot2

Scatterplots erlauben es uns Punkte entlang zweier Achsen zu visualisieren und so Korrelationen im Datansatz zu untersuchen. Wie dies mit ggplot geht schauen wir uns in dieser Lektion an.

Wir nutzen den bereits bekannten mtcars Datansatz:

```r
library('ggplot2')
df <- mtcars
```

```r
head(df)
```

## qplot()

```r
qplot(wt,mpg,data=df)
```

## Ein drittes Feature hinzufügen

Wir können durch einen Farbverlaub eine dritte Betrachtung bzw. Dimension zu jedem Punkt hinzufügen. Alternativ können wir die Größe der Punkte anhand dieses dritten Features anpassen. Zum Beispiel:

```r
qplot(wt,mpg,data=df,color=cyl)
```

```r
qplot(wt,mpg,data=df,size=cyl)
```

# Oder beides

```r
qplot(wt,mpg,data=df,size=gear,color=cyl)
```

```r
# Zeige 4 Dimensionen
qplot(wt,mpg,data=df,size=cyl,color=hp,alpha=0.6)
```

# ggplot()

Schauen wir uns nun an, wie wir durch ggplot() mehr Kontrolle erhalten:

```r
pl <- ggplot(data=df,aes(x = wt,y=mpg)) 
pl + geom_point()
```

## Ein drittes Feature hinzufügen

```r
pl <- ggplot(data=df,aes(x = wt,y=mpg)) 
pl + geom_point(aes(color=cyl))
```

```r
pl <- ggplot(data=df,aes(x = wt,y=mpg))

pl + geom_point(aes(color=factor(cyl)))
```

```r
pl <- ggplot(data=df,aes(x = wt,y=mpg))

pl + geom_point(aes(size=factor(cyl)))
```

```r
# Mit Formen
pl <- ggplot(data=df,aes(x = wt,y=mpg))

pl + geom_point(aes(shape=factor(cyl)))
```

```r
# Bessere Version
# mit Formen
pl <- ggplot(data=df,aes(x = wt,y=mpg))

pl + geom_point(aes(shape=factor(gear),color=factor(cyl)),size=4,alpha=0.6)
```

## Farbverläufe

```r
pl + geom_point(aes(colour = hp),size=4) + scale_colour_gradient(high='red',low = "blue")
```
