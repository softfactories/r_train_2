## Nutzung von ggplot()

Als nächstes schauen wir uns an, wie wir dies durch `ggplot` erweitern können! Die Syntax beginnt mit dem Grunddiagramm:

```r
# ggplot(Daten, Aesthetics)
pl <- ggplot(df,aes(x=rating))
```

```r
# Histogramm Geometry hinzufügen
pl + geom_histogram()
```

## Farbe hinzufügen

```r
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(binwidth=0.1,color='red',fill='pink') 
```

## Beschriftung hinzufügen

```r
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(binwidth=0.1,color='red',fill='pink') + xlab('Film Bewertungen')+ ylab('Erscheinungen') + ggtitle(' Film Bewertungen')
```

## Alpha (Transparenz) ändern

```r
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(binwidth=0.1,fill='blue',alpha=0.4) + xlab('Film Bewertungen')+ ylab('Erscheinungen')
```

## Linienarten

Wir haben folgende Optionen:"blank", "solid", "dashed", "dotted", "dotdash", "longdash", and "twodash". 

Bei einem Histogramm würden wir eigentlich keinen Gebrauch von anderen Linienartenmachen, können uns aber einmal ein Beispiel anschauen:

```r
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(binwidth=0.1,color='blue',fill='pink',linetype='dotted') + xlab('Film Bewertungen')+ ylab('Erscheinungen')
```

```r
# Beschriftungen hinzufügen
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(binwidth=0.1,aes(fill=..count..)) + xlab('Film Bewertungen')+ ylab('Erscheinungen')
```

You can further edit this by adding the **scale_fill_gradient()** function to your ggplot objects:

```r
# Beschriftung hinzufügen
pl <- ggplot(df,aes(x=rating))
pl2 <- pl + geom_histogram(binwidth=0.1,aes(fill=..count..)) + xlab('Film Bewertungen')+ ylab('Erscheinungen')
```

```r
# scale_fill_gradient('Label',low=color1,high=color2)
pl2 + scale_fill_gradient('Count',low='blue',high='red')
```

## Erweiterte Aesthetics

Wir können das `aes()` Argument an geom_histogram übergeben, um erweiterte Möglichkeiten der Gestalltung zu erhalten. Diese werden wir nicht im Detail betrachten, ein Beispiel sei aber behandelt: Es lassen sich Farbe und Füllungsskalen festlegen.

```r
# scale_fill_gradient('Label',low=color1,high=color2)
pl2 + scale_fill_gradient('Count',low='darkgreen',high='lightblue')
```

## Dichtediagramm hinzufügen

Wir können eine KDE ([kernel density estimation](https://en.wikipedia.org/wiki/Kernel_density_estimation)) zur Darstellung hinzufügen.

```r
# Beschriftung hinzufügen
pl <- ggplot(df,aes(x=rating))
pl + geom_histogram(aes(y=..density..)) + geom_density(color='red')
```
