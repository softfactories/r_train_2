## Lektion 58. R data.table

`data.table` ist eines der 15.000 Zusatzpakete für die beliebte Programmiersprache [R](https://www.r-project.org/about.html). Es bietet eine Hochleistungsversion von `data.frame` mit Syntax- und Funktionserweiterungen für Benutzerfreundlichkeit, Komfort und Programmiergeschwindigkeit. Ab November 2018 war data.table das viertgrößte Stack Overflow-Tag für ein R-Paket mit über [8.000 Fragen](http://stackoverflow.com/questions/tagged/data.table).

```r
require(data.table)
```

```r
DT = data.table(mtcars)
DT[1:6]
```

```r
# basic row subset
DT[2]                                       # 2nd row
DT[2:3]                                     # 2nd and 3rd row
w=2:3; DT[w]                                # same
# basic row subset

DT[2]                                       # 2nd row
DT[2:3]                                     # 2nd and 3rd row
w=2:3; DT[w]                                # same
```

```r
example(data.table) 
```
